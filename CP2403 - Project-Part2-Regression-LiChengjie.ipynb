# Import Libraries  

```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
import scipy
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
```

# Import Data

```python
bottle = pd.read_csv('bottle.csv')
bottle.head(1)
```

# Data management 

```python
bottle.info()
```

# Correlation between each explantory variable and response variable (y=total_cases)

```python
%matplotlib inline
Correlation1 = sns.regplot(x="Salnty", y="T_degC", fit_reg=False, data=bottle)
```

```python
%matplotlib inline
Correlation2 = sns.regplot(x="Oxy_µmol/Kg", y="T_degC", fit_reg=False, data=bottle)
```

# Scatter plot between each explantory variable and response variable (y=total_cases)

```python
%matplotlib inline
plt.figure()
scat2 = sns.regplot(x="Depthm", y="T_degC", fit_reg=False, data=bottle)
plt.xlabel('Depthm')
plt.ylabel('T_degC')
```

# Regression Analysis

```python
reg1 = smf.ols('Depthm ~ T_degC', data=bottle.head(100)).fit() 
print (reg1.summary())
```

# qq plot
